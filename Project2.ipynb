# 1. Bidang Terbanyak

```python
df['agency'].value_counts()
```

```python
terbanyak_bidang = df['agency'].value_counts().head()
terbanyak_bidang
```

```python
plt.figure(figsize=(34, 13))  # Mengatur ukuran figura
terbanyak_bidang.plot(kind='bar', color='skyblue')
plt.title('5 Bidang Terbanyak', fontsize=16)
plt.xlabel('Posisi', fontsize=14)
plt.ylabel('Frekuensi', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Menambahkan grid pada sumbu y
plt.show()
```

Untuk CPNS 2024 bidang paling banyak dibuka itu ada di Kementrian Pekerjaan Umum dan Perumahan Rakyat di ikuti dengan Kementerian Agama kemudian Pemerintah Provinsi Daerah Khusus Ibukota Jakarta lalu ada Kementerian Pendidikan, Kebudayaan, Riset dan Teknologi dan terkahir adalah Kementerian Kesehatan.

# 2. Posisi Terbanyak

```python
df['position'].value_counts()
```

```python
terbanyak_posisi = df['position'].value_counts().head()
terbanyak_posisi
```

```python
plt.figure(figsize=(28, 8))  # Mengatur ukuran figura
terbanyak_posisi.plot(kind='bar', color='skyblue')
plt.title('5 Posisi Terbanyak', fontsize=16)
plt.xlabel('Posisi', fontsize=14)
plt.ylabel('Frekuensi', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Menambahkan grid pada sumbu y
plt.show()
```

Untuk CPNS tahun 2024 posisi terbanyak itu di posisi ADMINISTRATOR KESEHATAN AHLI PERTAMA dengan jumlah posisi pekerjaan yang dibuka sebanyak 50400 diikuti dengen posisi PENATA KELOLA SISTEM DAN TEKNOLOGI INFORMASI kemudian PERENCANA AHLI PERTAMA lalu ARSIPARIS AHLI PERTAMAdan terakhir di posisi PRANATA KOMPUTER AHLI PRATAMA

# 3. Tingkat Pendidikan Terbanyak

```python
education = df['education_level'].value_counts()
education
```

```python
plt.figure(figsize=(28, 8))  # Mengatur ukuran figura
education.plot(kind='bar', color='skyblue')
plt.title('Education', fontsize=16)
plt.xlabel('Education', fontsize=18)
plt.ylabel('Frekuensi', fontsize=18)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Menambahkan grid pada sumbu y
plt.show()
```

**Tahun 2024 tingkat pendidikan paling banyak dibuka yaitu untuk lulusan S-1/ Sarjana kemudian D3 lanjut D4 dan S2**
